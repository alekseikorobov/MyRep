## 1. Создание репозитория в GitHub

Для начала нужно создать репозиторий в GitHub\
Авторизируемся под своим акаунтом\
Выбираем **New repository**

#### Задаем имя и все остальные настройки оставляем по умолчанию

Нажимаем **Create Repository**

Дальше выходит инструкция, чтобы можно было получить репозиторий к себе

Для нас подходит инструкция по первому сценарию, так как еще нет репозитория

## 2. Получения репозитория

Запускаем команды как предлагается в инструкции, только можно без первой строчки и мы сразу добавим в репозиторий все файлы
И так как мы будем делать это в notebook, то перед git нужно поставить !
```
!git init
!git add .
!git commit -m "first commit"
!git branch -M main
!git remote add origin https://github.com/alekseikorobov/MyRep.git
!git push -u origin main
```
